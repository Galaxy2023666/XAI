```python
import pandas as pd
import json
import re
import simplejson  # 更健壮的JSON解析库

def extract_title_url_pairs(assistant_data: dict) -> list:
    """从assistant数据中提取title和url对"""
    results = []
    # 确保searchGuid存在且是列表
    search_guid = assistant_data.get('searchGuid', [])
    if not isinstance(search_guid, list):
        return results
    
    for guid in search_guid:
        # 确保guid是字典类型
        if isinstance(guid, dict) and 'docs' in guid:
            for doc in guid['docs']:
                # 确保doc是字典类型
                if isinstance(doc, dict):
                    results.append({
                        "title": doc.get('title', ''),
                        "url": doc.get('url', '')
                    })
    return results

def fix_json_errors(json_str: str) -> str:
    """修复常见的JSON格式错误"""
    if not json_str or not isinstance(json_str, str):
        return '[]'
    
    # 1. 添加缺失的数组括号
    if not json_str.strip().startswith('['):
        json_str = '[' + json_str
    if not json_str.strip().endswith(']'):
        json_str += ']'
    
    # 2. 修复未终止的字符串
    json_str = re.sub(r'([^\\]")', r'\1', json_str)  # 确保每个"都有匹配
    
    # 3. 转义特殊字符
    json_str = json_str.replace('\n', '\\n').replace('\r', '\\r')
    
    # 4. 修复未转义的控制字符
    json_str = re.sub(r'[\x00-\x1F]', ' ', json_str)
    
    return json_str

def process_model_content(json_data: str) -> tuple:
    """处理模型A问答_content，返回pairs和extracted_pairs"""
    # 空值处理
    if not json_data or not isinstance(json_data, str) or json_data.strip() == "":
        return [], []
    
    try:
        # 先修复JSON格式问题
        fixed_json = fix_json_errors(json_data)
        
        # 使用simplejson解析（更容错）
        parsed = simplejson.loads(fixed_json, strict=False)
        
        pairs = []
        extracted_pairs_list = []
        
        # 创建user-assistant对
        for i in range(0, len(parsed), 2):
            if i + 1 < len(parsed):
                user_msg = parsed[i]
                assistant_msg = parsed[i + 1]
                
                # 确保角色字段存在
                if (isinstance(user_msg, dict) and isinstance(assistant_msg, dict) and
                    user_msg.get('role') == 'user' and assistant_msg.get('role') == 'assistant'):
                    
                    # 创建对话对文本
                    pair_text = f"用户: {user_msg.get('content', '')}\n助手: {assistant_msg.get('content', '')}"
                    pairs.append(pair_text)
                    
                    # 提取文档对
                    extracted = extract_title_url_pairs(assistant_msg)
                    extracted_pairs_list.append(extracted)
        
        return pairs, extracted_pairs_list
    
    except Exception as e:
        print(f"JSON处理错误: {str(e)[:200]}")
        print(f"问题数据前100字符: {json_data[:100]}")
        return [], []

def process_excel(input_path: str, output_path: str):
    """处理Excel文件主函数"""
    # 读取Excel文件，指定文本类型防止截断
    df = pd.read_excel(
        input_path,
        engine='openpyxl',
        dtype={'用户Prompt': str, '模型A问答': str}
    )
    
    # 添加临时组ID标识原始行
    df['_group_id'] = df.index
    
    # 拆分用户Prompt列
    df['用户Prompt'] = df['用户Prompt'].str.split('--------问答轮次分割线--------')
    
    # 创建错误日志容器
    error_log = []
    
    def safe_processor(x):
        try:
            return process_model_content(x)
        except Exception as e:
            error_log.append({
                '原始数据': str(x)[:500],  # 截取前500字符
                '错误信息': str(e)
            })
            return [], []
    
    # 处理模型A问答列
    df[['pairs', 'extracted_pairs']] = df['模型A问答'].apply(
        lambda x: pd.Series(safe_processor(x))
    )
    
    # 验证拆分一致性
    validation_passed = df.apply(
        lambda x: len(x['用户Prompt']) == len(x['pairs']) == len(x['extracted_pairs']), 
        axis=1
    )
    
    if not all(validation_passed):
        print(f"警告: {len(validation_passed) - sum(validation_passed)} 行数据拆分不一致")
    
    # 同时爆炸多列生成多行
    df_exploded = df.explode(['用户Prompt', 'pairs', 'extracted_pairs'])
    
    # 标记每组的第一行
    df_exploded['_is_first'] = df_exploded.groupby('_group_id').cumcount() == 0
    
    # 处理其他列：仅第一行保留原值
    other_columns = [col for col in df.columns if col not in [
        '用户Prompt', 'pairs', 'extracted_pairs', '模型A问答', '_group_id'
    ]]
    
    for col in other_columns:
        df_exploded[col] = df_exploded.apply(
            lambda row: row[col] if row['_is_first'] else '', 
            axis=1
        )
    
    # 删除临时列
    df_exploded.drop(columns=['_group_id', '_is_first', '模型A问答'], inplace=True)
    
    # 保存处理结果
    df_exploded.to_excel(output_path, index=False, engine='openpyxl')
    
    # 保存错误日志
    if error_log:
        error_df = pd.DataFrame(error_log)
        error_path = output_path.replace('.xlsx', '_errors.xlsx')
        error_df.to_excel(error_path, index=False, engine='openpyxl')
        print(f"保存了 {len(error_log)} 条错误记录到: {error_path}")

# 使用示例
process_excel("/Users/serena/Downloads/work/多轮_deepseek_30session_线上原始数据.xlsx", "/Users/serena/Downloads/work/sheet1.xlsx")
```

```python
import pandas as pd
import json
from typing import List, Dict, Tuple

def extract_title_url_pairs(assistant_data: dict) -> List[Dict[str, str]]:
    """从assistant数据中提取title和url对"""
    output = ""  # 创建一个空字符串用于存储结果
    if 'searchGuid' not in assistant_data:
        return output
    
    for guid in assistant_data['searchGuid']:
        if 'docs' in guid and guid['docs'] is not None: 
            for doc in guid['docs']:
                title = doc.get('title', '')
                url = doc.get('url', '')
            # 将键值对格式化为字符串，并在末尾添加换行符
                output += f"{title}, {url}\n"  

    return output

def process_model_content(json_data: str) -> Tuple[List[str], List[List[Dict]]]:
    """处理模型问答_content，返回pairs和extracted_pairs"""
    if isinstance(json_data, (float, int)):  # 如果是数字类型（包括float类型）
        print(f"警告: 发现非字符串数据: {json_data}，跳过此项.")
        return [], []  # 跳过该项，返回空列表

    if isinstance(json_data, str):  # 确保传入的是字符串
        try:
            parsed = json.loads(json_data)  # 解析JSON数据
        except json.JSONDecodeError as e:
            print(f"错误: 无法解析JSON: {json_data}\n{e}")
            return [], []  # 返回空列表，跳过此项
    else:
        print(f"错误: 发现无效数据类型: {type(json_data)}，跳过此项.")
        return [], []  # 返回空列表，跳过此项
    
    parsed = json.loads(json_data)
    pairs = []
    extracted_pairs_list = []
    
    # 创建user-assistant对
    for i in range(0, len(parsed), 2):
        if i + 1 < len(parsed):
            user_msg = parsed[i]
            assistant_msg = parsed[i + 1]
            
            if user_msg['role'] == 'user' and assistant_msg['role'] == 'assistant':
                # 创建对话对文本
                pair_text = f"用户: {user_msg}\n助手: {assistant_msg}"
                pairs.append(pair_text)
                
                # 提取文档对
                extracted = extract_title_url_pairs(assistant_msg)
                extracted_pairs_list.append(extracted)
    
    return pairs, extracted_pairs_list

def process_csv(input_file: str, output_file: str):
    """处理CSV文件主函数"""
    # 读取CSV文件
    df = pd.read_csv(input_file)
    
    # 添加临时组ID标识原始行
    df['_group_id'] = df.index
    
    # 拆分用户Prompt列
    df['用户Prompt'] = df['用户Prompt'].str.split('--------问答轮次分割线--------')
    
    # 处理模型A问答_content列并将拆分列添加到相应位置
    for i in range(len(df)):
        x = df['模型A问答'].iloc[i]
        df[['pairs_A', 'url_pairs_A']] = df['模型A问答'].apply(
            lambda x: pd.Series(process_model_content(x))
        )
        
    # 处理模型B问答_content列并将拆分列添加到相应位置
    for i in range(len(df)):
        y = df['模型B问答'].iloc[i]
        df[['pairs_B', 'url_pairs_B']] = df['模型B问答'].apply(
            lambda y: pd.Series(process_model_content(y))
        )
    
    # 同时爆炸多列生成多行
    df_exploded = df.explode(['用户Prompt', 'pairs_A', 'url_pairs_A', 'pairs_B', 'url_pairs_B'])
    
    # 标记每组的第一行
    df_exploded['_is_first'] = df_exploded.groupby('_group_id').cumcount() == 0
    
    # 处理其他列：仅第一行保留原值
    other_columns = [col for col in df.columns if col not in [
        '用户Prompt', 'pairs_A', 'url_pairs_A', 'pairs_B', 'url_pairs_B', '_group_id'
    ]]
    
    for col in other_columns:
        df_exploded[col] = df_exploded.apply(
            lambda row: row[col] if row['_is_first'] else '', 
            axis=1
        )
    
    # 删除临时列并保存结果
    df_exploded.drop(columns=['_group_id', '_is_first', 'pairs_A', 'pairs_B'], inplace=True)
    df_exploded.to_csv(output_file, index=False, encoding='utf-8-sig')

# 使用示例
process_csv("/Users/serena/Downloads/work/model_run_resul_merge.csv", "/Users/serena/Downloads/work/model_run_resul_merge_new.csv")
```
